# Loading the trained model

```python
model = tf.keras.models.load_model('/content/drive/MyDrive/Final Acne Analyzer/Final Model')
class_names = ['1-Clear', '2-Mild', '3-Moderate', '4-Severe', '5-Very_Severe']
```

# Passing skin patches through the model

```python
imgs = [forehead_region, chin_region, right_cheek, left_cheek]

img_height = 240
img_width = 240
sum_acne = 0

for image in imgs:
  img = keras.preprocessing.image.load_img(
      image, target_size=(img_height, img_width))
  img_array = keras.preprocessing.image.img_to_array(img)
  img_array = tf.expand_dims(img_array, 0) # Create a batch

  predictions = model.predict(img_array)
  score = tf.nn.softmax(predictions[0])

  print(
      "This patch most likely belongs to {} category."
      .format(class_names[np.argmax(score)])
  )

  # Calculating overall acne:
  overall_acne = class_names[np.argmax(score)]
  overall_acne = int(overall_acne[0])
  sum_acne = sum_acne + overall_acne
  avg_acne = round((sum_acne / 4) - 1)

print('You have '+ class_names[avg_acne]+ ' acne.')
```
